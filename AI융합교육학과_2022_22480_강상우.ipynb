#이미지 업로드

```python
#@markdown ### 업로드할 이미지를 선택하여 업로드해 주세요.

import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

#(1)file upload
filename = list(files.upload().keys())[0]
```

#이미지

```python
#@markdown ### 원본이미지

#(2)upload image view
img = cv2.imread(filename)
cv2_imshow(img)
```

#확대, 축소

```python
#@markdown ### 확대 또는 축소할 가로, 세로 비율을 적어 주세요.

x_scale = 0.1 #@param {type:"number"}
y_scale = 0.1 #@param {type:"number"}

img_scale = cv2.resize(img, None, fx=x_scale, fy=y_scale)
cv2_imshow(img_scale)
```

#회전

```python
#@markdown ### 반시계방향으로 이미지 회전할 각도를 적어 주세요.
import numpy as np

img = cv2.imread(filename)
w, h = img.shape[:2]

rotation_degree =  60#@param {type:"number"}
# rotation_degree = np.mod(rotation_degree, 90)
rotation_radian = rotation_degree / 180 * np.pi

sin = np.sin(rotation_radian)
cos = np.cos(rotation_radian)
w_new = w * cos + h * sin
w_new = np.abs(w_new)
h_new = w * sin + h * cos
h_new = np.abs(h_new)

scale = min(w / w_new, h / h_new)

rotation = cv2.getRotationMatrix2D((h/2, w/2), rotation_degree, scale)
img_rotation = cv2.warpAffine(img, rotation, (h, w), borderValue=(255, 255, 255))
cv2_imshow(img_rotation)
```

#평행이동

```python
#@markdown ### 평행이동할 row값과 col값을 적어 주세요.
import numpy as np

img = cv2.imread(filename)
col, row = img.shape[:2]

move_row = 100 #@param {type:"number"}
move_col = 100 #@param {type:"number"}

translation = np.float32([[1,0,move_row],
                          [0,1,move_col]])
img_translation = cv2.warpAffine(img, translation, (row, col), borderValue=(255, 255, 255))
cv2_imshow(img_translation)
```

#합성

```python
#@markdown ### 이미지가 겹쳐지는 효과를 내기 위해서 이미지 합성을 할 이미지를 업로드해 주세요.
from cv2 import imread
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from google.colab import files
from IPython.display import HTML
from tqdm import tqdm_notebook

#print("Upload two files to cross dissolve")


#(1)file upload
filename2 = list(files.upload().keys())[0]

#(2)upload image view
img2 = cv2.imread(filename2)
cv2_imshow(img2)
```

```python
#@markdown ### 스크롤을 움직이며 이미지 겹치는 효과를 확인하세요.
i1 = imread(filename) / 256
i2 = imread(filename2) / 256

i1 = np.flip(i1, axis=2)
i2 = np.flip(i2, axis=2)

h, w, _ = i2.shape
i1 = cv2.resize(i1, dsize=((w, h)))
#print(w, h)

fig = plt.figure()
plt.axis('off')

ims = []
MAX = 50

for i in tqdm_notebook(range(MAX + 1)):
    im = plt.imshow(i2 * (i / MAX) + i1 * (1 - i / MAX), animated=True)
    ims.append([im])

plt.close()
ani = animation.ArtistAnimation(fig, ims, interval=30, blit=True)
HTML(ani.to_jshtml())
```
